```python
import os
import pytesseract
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# -------- STEP 1: Load Text --------

with open("extracted_text/model_text.txt", "r", encoding="utf-8") as f:
    model_text = f.read()

with open("extracted_text/student_text.txt", "r", encoding="utf-8") as f:
    student_text = f.read()

# -------- STEP 2: Clean Text --------

def clean_text(text):
    text = text.lower()
    text = text.replace("\n", " ")
    return text

model_text = clean_text(model_text)
student_text = clean_text(student_text)

# Debug check
print("Model Text Sample:", model_text[:200])
print("Student Text Sample:", student_text[:200])

# -------- STEP 3: TF-IDF + Cosine Similarity --------

vectorizer = TfidfVectorizer(stop_words='english')

texts = [model_text, student_text]
tfidf_matrix = vectorizer.fit_transform(texts)

similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]

print("\n✅ Similarity Score:", round(similarity, 3))

# -------- STEP 4: Convert to Marks --------

max_marks = 20   # 👈 change based on your paper

obtained_marks = round(similarity * max_marks, 2)

# -------- STEP 5: Result --------

print("\n----------- FINAL RESULT -----------")
print("Marks Obtained:", obtained_marks)
print("Out of:", max_marks)

percentage = (obtained_marks / max_marks) * 100
print("Percentage:", round(percentage, 2), "%")
```
