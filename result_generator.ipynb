```python
import re
import csv
import os
```

```python
with open("processed/model/text/model.txt", "r") as f:
    model_text = f.read()

with open("processed/students/student_1/text/student.txt", "r") as f:
    student_text = f.read()

print("Files Loaded Successfully")
print("Model Text Sample:\n", model_text[:300])
print("\nStudent Text Sample:\n", student_text[:300])
```

```python
def remove_header(text):
    if "1." in text:
        return text.split("1.", 1)[1]
    return text

model_text = remove_header(model_text)
student_text = remove_header(student_text)
```

```python
def clean_text(text):
    # Fix merged cases like Answer:B2.
    text = re.sub(r'([A-D])(\d+\.)', r'\1\n\2', text)
    
    # Fix OCR issue like BQ.
    text = re.sub(r'([A-D])Q\.', r'\1\nQ.', text)
    
    # Ensure space after number
    text = re.sub(r'(\d+)\.', r'\1. ', text)
    
    return text

model_text = clean_text(model_text)
student_text = clean_text(student_text)
```

```python
def split_mcqs(text):
    questions = re.split(r'\n(?=\d+\.)', text)
    return [q.strip() for q in questions if q.strip()]

model_mcqs = split_mcqs(model_text)
student_mcqs = split_mcqs(student_text)

print("\nModel Questions:", len(model_mcqs))
print("Student Questions:", len(student_mcqs))
```

```python
def extract_option(text):
    text = text.upper()
    
    # Fix OCR mistakes
    text = text.replace("8", "B").replace("0", "D")
    
    match = re.search(r'\b([A-D])\b', text)
    return match.group(1) if match else ""
```

```python
score = 0
total = min(len(model_mcqs), len(student_mcqs))

print("\nQuestion-wise Result:\n")

for i in range(total):
    
    model_ans = extract_option(model_mcqs[i])
    student_ans = extract_option(student_mcqs[i])
    
    if model_ans == student_ans and model_ans != "":
        score += 1
        print(f"Q{i+1}: ✅ Correct ({model_ans})")
    else:
        print(f"Q{i+1}: ❌ Wrong (Model: {model_ans}, Student: {student_ans})")
```

```python
percentage = (score / total) * 100

print("\n📊 FINAL RESULT")
print("Total Marks:", score)
print("Out of:", total)
print("Percentage:", round(percentage, 2), "%")
```

```python
file_path = "results/results.csv"

file_exists = os.path.exists(file_path) and os.path.getsize(file_path) > 0

with open(file_path, "a", newline="") as f:
    writer = csv.writer(f)
    
    if not file_exists:
        writer.writerow(["Student_ID", "Marks", "Total", "Percentage"])
    
    writer.writerow(["student_1", score, total, round(percentage, 2)])

print("\n Result saved to results/results.csv")
```
