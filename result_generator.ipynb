```python
# If similarity notebook not run, recreate marks here

import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load texts
with open("extracted_text/model_text.txt", "r") as f:
    model_text = f.read()

with open("extracted_text/student_text.txt", "r") as f:
    student_text = f.read()

# Split function
def split_answers(text):
    parts = re.split(r'(Q\d+)', text)
    answers = {}
    for i in range(1, len(parts), 2):
        answers[parts[i]] = parts[i+1].strip()
    return answers

model_answers = split_answers(model_text)
student_answers = split_answers(student_text)

# Calculate similarity and marks
marks = {}
max_marks_per_question = 5

for question in model_answers:
    if question in student_answers:
        vectorizer = TfidfVectorizer()
        texts = [model_answers[question], student_answers[question]]
        tfidf_matrix = vectorizer.fit_transform(texts)
        similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
        marks[question] = round(similarity * max_marks_per_question, 2)

# Now calculate total
total_marks = sum(marks.values())
max_total = len(marks) * max_marks_per_question

print("----------- RESULT SUMMARY -----------")
print("Total Marks Obtained:", round(total_marks, 2))
print("Out of:", max_total)

if max_total != 0:
    percentage = (total_marks / max_total) * 100
else:
    percentage = 0

print("Percentage:", round(percentage, 2), "%")
```
